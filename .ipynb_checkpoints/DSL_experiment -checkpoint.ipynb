```python
def create_matrix():
        print("Enter number of rows and columns for the matrix -> space seperated")
        row ,col = map(int,input().split())
        print("Enter elements in matrix representation")
        print ("for example for 2X2 matrix if elements are a,b,c,d")
        print("format = ")
        print("a b")
        print("c d")
        matrix = list()
        for i in range(row):
            matrix.append(list())
            for j in input().split():
                matrix[-1].append(int(j))
            return matrix
class matrix:
    
    def add(first,second):
        if len(first)!=len(second) or len(first[-1]) != len(second[-1]):
            print("Matrices cannot be added!! check order of matrices before adding")
            return 
        else:
            matrix = list()
            for i in range(len(first[0])):
                matrix.append(list())
            for i in range(len(first)):
                for j in range(len(first[0])):
                    matrix[i].append(first[i][j] + second[i][j])
            if len(matrix[-1]) == 0:
                matrix.pop()
            return matrix
    def substract(first,second):
        if len(first)!=len(second) or len(first[-1]) != len(second[-1]):
            print("Matrices cannot be added!! check order of matrices before substracting")
            return 
        else:
            matrix = list()
            for i in range(len(first[0])):
                matrix.append(list())
            for i in range(len(first)):
                for j in range(len(first[0])):
                    matrix[i].append(first[i][j] - second[i][j])
            if len(matrix[-1]) == 0:
                matrix.pop()
            return matrix
    def multiply(first,second):
        if len(first[0]) != len(second):
            print("matrices cannot be multiplied")
            return 
        else:
            matrix = list()
            for i in range(len(first)):
                matrix.append(list())
                for j in range(len(second[0])):
                    matrix[-1].append(0)
            for i in range(len(first)):
                for j in range(len(second[0])):
                    for k in range(len(second)):
                        matrix[i][j] += first[i][k]*second[k][j]
            return matrix
    def transpose(first):
        matrix = list()
        r = len(first)
        c = len(first[0])
        for i in range(c):
            matrix.append(list())
            for j in range(r):
                matrix[-1].append(0)
        for i in range(r):
            for j in range(c):
                matrix[j][i] = first[i][j]
        return matrix

print("Enter the values of first matrix")
a = Matrix.create_matrix()
print("Enter the values of second matrix")
b = Matrix.create_matrix()

print('press y to continue giving queries , no to stop the program')
c = input('Enter your choice')
while(c == 'y' or  c == 'Y'):
    print('press 1 to get the sum of two matrices')
    print('press 2 to get the difference of two matrices')
    print('press 3 to get the product of two matrices')
    print('press 4 to get the transpose of first matrix')
    print('press 5 to get the transpose of second matrix')
    q = int(input("Enter your choice"))
    if q == 1:
      ans = Matrix.add(a,b)
      for i in range(ans):
        for j in range(ans[0]):
          print(ans[i][j] , end = " ")
        print()
    if q == 2:
      ans = Matrix.substract(a,b)
      for i in range(ans):
        for j in range(ans[0]):
          print(ans[i][j] , end = " ")
        print()
    if q == 3:
      ans = Matrix.multiply(a,b)
      for i in range(ans):
        for j in range(ans[0]):
          print(ans[i][j] , end = " ")
        print()
    if q == 4:
      ans = Matrix.transpose(a)
      for i in range(ans):
        for j in range(ans[0]):
          print(ans[i][j] , end = " ")
        print()
    if q == 5:
      ans = Matrix.transpose(b)
      for i in range(ans):
        for j in range(ans[0]):
          print(ans[i][j] , end = " ")
        print()
    inp = input("do you want to continue , press y to continue else n to exit")  
    c = inp
    
```
